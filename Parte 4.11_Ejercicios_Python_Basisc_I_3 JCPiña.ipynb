## Ejercicio 1
Tenemos las siguientes variables:

```Python
A = 4
B = "Text"
C = 4.1
```

Comprueba:
1. Si A y B son equivalentes
2. Si A y C NO son equivalentes
3. Si A es mayor que C 
4. Si C es menor o igual que A
5. Si B NO es equivalente a C

````python
A = 4
B = "Text"
C = 4.1
print (A == B)
print (A != C)
print (A > C)
print (C <= A)
print (B != C)
````

## Ejercicio 3
Mismo programa que en 2, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
signo_zodiacal1 = str(input("signo zodiacal persona 1"))
signo_zodiacal2 = str(input("signo zodiacal persona 2"))
signo_zodiacal3 = str(input("signo zodiacal persona 3"))
if signo_zodiacal1 == signo_zodiacal2 == signo_zodiacal3:
    print ("Sois del mismo signo zodiacal")
elif signo_zodiacal1 == signo_zodiacal2 != signo_zodiacal3 or signo_zodiacal1 != signo_zodiacal2 == signo_zodiacal3 or signo_zodiacal1 == signo_zodiacal3 != signo_zodiacal2:
    print ("Dos de vosotros tenéis el mismo signo zodiacal")
else:
    print("Ninguno tenéis el mismo signo zodiacal")
````
