### Ejercicio 3

Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

```python
# establecer los input
base = float (input ("indica la base"))
altura = float (input ("indica la altura"))
```

```python
# presentar resultado
print (area)
print ("el area de este paralelogramo con una base", base , "y con una altura", altura, "es:",area)
```

### Ejercicio 4

Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`. En este caso debe ser capaz de admitir valores reales, con decimales, para base y altura en español. (PISTA: Recuerda que tendrás que hacer uso del método replace)

```python
# establecer los input
base = (input ("indica la base"))
altura = (input ("indica la altura"))
```

```python
# indicar formula y remplazar valores a decimales
area = float(base.replace (",","."))* float(altura.replace (",","."))
```

```python
# presentar resultado
print (area)
print ("el area de este paralelogramo con una base", base , "y con una altura", altura, "es:",area)
```
